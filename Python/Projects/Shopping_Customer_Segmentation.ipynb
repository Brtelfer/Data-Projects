```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame(pd.read_csv('/Users/brtelfer/Documents/Python_Data_Projects/16_Shopping_Customer_Segmentation/Mall_Customers.csv'))
```

```python
sns.distplot(df['Annual Income (k$)']);
```

```python
col_lst = df.columns.to_list()
col_list = col_lst.remove('CustomerID')
for x in col_lst:
    plt.figure()
    sns.displot(df[x])
    plt.tight_layout()
    plt.show()
```

```python
col_lst = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in col_lst:
    plt.figure()
    sns.kdeplot(data=df, x=i, hue='Gender', shade=True, common_norm=False, palette="crest", alpha=.3, linewidth=2);
    plt.title(f'KDE plot for {i}')
```

```python
col_lst = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in col_lst:
    plt.figure()
    sns.boxplot(data=df, x='Gender',y=df[i],palette='Set1');
    plt.title(f'Boxplot by {i}')
```
