```python
df = df_1.copy()
```

```python
df['Used name'] = df['Used name'].str.replace("•"," ")
```

```python
df[['Height_cm','Weight_kg']] = df['Measurements'].str.split('/', expand=True)
```

```python
df['Height_cm'] = df['Height_cm'].str.replace('cm',' ')
df['Weight_kg'] = df['Weight_kg'].str.replace('kg',' ')
```

```python
small_data = pd.DataFrame(['183 cm', '180 cm / 150 kg', '190 cm / 80 kg', '100 kg'], columns=['Measurements'])
small_data[['Height_cm','Weight_kg']] = small_data['Measurements'].str.split('/', expand=True)
small_data['Weight_kg'].fillna(small_data['Height_cm'], inplace=True)
small_data['Weight_kg'] = np.where(small_data['Weight_kg'].str.contains("cm"), np.nan, small_data['Weight_kg'])
small_data['Height_cm'] = np.where(small_data['Height_cm'].str.contains("kg"), np.nan, small_data['Height_cm'])
small_data.head()
```

```python
df['Weight_kg'] = np.where(df['Weight_kg'].str.contains("cm"), np.nan, df['Weight_kg'])
df['Height_cm'] = np.where(df['Height_cm'].str.contains("kg"), np.nan, df['Height_cm'])
```
